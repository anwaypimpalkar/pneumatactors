```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the CSV file
file_path = 'Data/Logs/U01_Trial01_20250610_195247.csv'
df = pd.read_csv(file_path)

# Display column names to identify proxy x, y, z
df.columns

# Create a 3D plot for proxyThumb and proxyIndex trajectories
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot proxy thumb
ax.plot(df['proxyThumb_x'], df['proxyThumb_y'], df['proxyThumb_z'], label='Proxy Thumb', linewidth=1)

# Plot proxy index
ax.plot(df['proxyIndex_x'], df['proxyIndex_y'], df['proxyIndex_z'], label='Proxy Index', linewidth=1)

# Labels and legend
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Proxy Thumb and Index Trajectories in 3D Space')
ax.legend()

plt.tight_layout()
plt.show()
```
