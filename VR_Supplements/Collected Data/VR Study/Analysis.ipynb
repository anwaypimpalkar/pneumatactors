```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob

# --- Folder with your trial files ---
folder_path = 'QuestLogs/Logs/U02'
file_pattern = os.path.join(folder_path, '*.csv')

summary = []

for file_path in sorted(glob.glob(file_pattern)):
    try:
        df = pd.read_csv(file_path)

        # Check for required columns
        if 'TrialTime' not in df.columns or 'broken' not in df.columns:
            print(f"[Skipped] Missing columns in: {file_path}")
            continue

        df['broken'] = df['broken'].astype(int)

        # Detect rising edges (0 → 1)
        df['broken_shifted'] = df['broken'].shift(fill_value=0)
        df['break_event'] = (df['broken'] == 1) & (df['broken_shifted'] == 0)
        break_event_count = df['break_event'].sum()

        # Save summary
        summary.append({
            "File": os.path.basename(file_path),
            "Break Events": break_event_count
        })

        # Optional: plot each file (comment out to skip)
        plt.figure(figsize=(10, 3))
        plt.plot(df['TrialTime'], df['broken'], label='Broken State')
        plt.title(f"Break Events - {os.path.basename(file_path)}")
        plt.xlabel("Time (s)")
        plt.ylabel("Broken")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"[Error] Failed to process {file_path}: {e}")

# Print summary
summary_df = pd.DataFrame(summary)
print("\n===== Break Event Summary =====")
print(summary_df)

# Save summary to CSV
summary_df.to_csv("break_summary_all_trials.csv", index=False)
```
