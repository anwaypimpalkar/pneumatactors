```python
import random
import json
import os
import pandas as pd

random.seed(19)  # For reproducibility

def generate_trials(user_id="u01"):
    user_number = int(user_id[1:])
    pressure_levels = [1, 2, 3, 4]
    vibration_levels = [1, 2, 3, 4]

    num_blocks_pressure = 5
    num_blocks_vibration = 5
    num_blocks_multimodal = 3
    blocks = []

    # Determine trial order
    if user_number % 2 == 1:
        first, second = "vibration", "pressure"
    else:
        first, second = "pressure", "vibration"
    trial_order = [first, second, "multimodal"]

    for trial_type in trial_order:
        if trial_type == "pressure":
            for _ in range(num_blocks_pressure):
                block = [
                    {
                        "trial_type": "pressure",
                        "pressure_level": p,
                        "vibration_level": 0
                    } for p in pressure_levels
                ]
                random.shuffle(block)
                blocks.extend(block)

        elif trial_type == "vibration":
            for _ in range(num_blocks_vibration):
                block = [
                    {
                        "trial_type": "vibration",
                        "pressure_level": 3,
                        "vibration_level": v
                    } for v in vibration_levels
                ]
                random.shuffle(block)
                blocks.extend(block)

        elif trial_type == "multimodal":
            for _ in range(num_blocks_multimodal):
                block = [
                    {
                        "trial_type": "multimodal",
                        "pressure_level": p,
                        "vibration_level": v
                    } for p in pressure_levels for v in vibration_levels
                ]
                random.shuffle(block)
                blocks.extend(block)

    # Now assign correct sequential trial numbers
    for idx, trial in enumerate(blocks, start=1):
        trial["trial_number"] = idx

    output_dir = "3exp_trials_counterbalanced"
    os.makedirs(output_dir, exist_ok=True)
    filename = os.path.join(output_dir, f"{user_id}_trials.json")
    with open(filename, "w") as f:
        json.dump(blocks, f, indent=4)
    print(f"✅ Saved to: {filename}")

# Create and save trial files
user_ids = [f"u{i:02}" for i in range(1, 25)]
for user_id in user_ids:
    generate_trials(user_id)

# # Create a DataFrame for display
# filenames = [f"{user_id}_trials.json" for user_id in user_ids]
# df = pd.DataFrame({"User": user_ids, "Filename": filenames})
# print(df)
```
